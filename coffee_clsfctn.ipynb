```python
import pandas as pd
```

```python
data = pd.read_csv("coffee.csv")
data.head(2)
```

```python
dataset = data[["Country of Origin","Variety","Aroma","Flavor","Aftertaste","Acidity","Body","Balance","Moisture Percentage","Total Cup Points"]]
dataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder

dataset["Country of Origin"] = LabelEncoder().fit_transform(dataset["Country of Origin"].values)
dataset["Variety"] = LabelEncoder().fit_transform(dataset["Variety"].values)

dataset.head()
```

```python
dataset.isnull().value_counts()
```

```python
dataset["Country of Origin"].value_counts()
```

```python
dataset.count()
```

```python
lst = [2,13,3,17,20,11,15,21,19,7,9,14,8,10,12]
for i in range(0,207):
  if dataset.loc[i,"Country of Origin"] in lst:
      dataset = dataset.drop(i)

dataset["Country of Origin"].value_counts()
```

```python
dataset.count()
```

```python
from sklearn.preprocessing import LabelEncoder

dataset["Country of Origin"] = LabelEncoder().fit_transform(dataset["Country of Origin"].values)
dataset['Country of Origin'].value_counts()
```

```python
x = dataset.drop('Country of Origin',axis=1)
y = dataset['Country of Origin']

x
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.4,random_state=42)
```

```python
x_train.count()
x_test.count()
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=10000)
```

```python
model.fit(x_train, y_train)

predicted = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,predicted)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,predicted))
```

```python
from sklearn.svm import LinearSVC

model2 = LinearSVC()

model2.fit(x_train,y_train)
predicted2 = model2.predict(x_test)

predicted2.size
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,predicted2))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

model3 = KNeighborsClassifier(n_neighbors=9)

model3.fit(x_train,y_train)
predicted3 = model3.predict(x_test)

print(classification_report(y_test,predicted3))
```
